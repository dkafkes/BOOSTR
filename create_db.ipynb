```
for param in ['Event0c_complete']:
  if re.search(':', param) is not None:
    param2 = param.replace(":", "")
  else:
    param2 = param
  
  fpath = "/mnt/gmps/"+str(param)
  save_where = "/mnt/delta/"+str(param)
  data = spark.read.parquet(fpath)
  data2 = data.select(col("filename").alias("file"), col("time"), col("SCmicrosec"), col("UNIXepoch"))
  data2.write.partitionBy("file").format("delta").mode("overwrite").option("overwriteSchema", "true").save(save_where)

  command = ''' CREATE TABLE {} USING DELTA LOCATION '{}';'''.format(param2, save_where)
  
  print(command)
  spark.sql(command)
  print("done with "+str(param))
```
