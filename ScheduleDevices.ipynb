```
%scala
import scala.concurrent.ExecutionContext.Implicits.global
import scala.concurrent.duration._
import scala.language.postfixOps
import scala.concurrent.{Future, Await}
import scala.concurrent.duration._
import scala.util.control.NonFatal

val booster_lags = spark.read.format("csv").option("header", "true").load("/FileStore/tables/booster_lags/last_try.csv")
val params = booster_lags.select("param").as[String].collect()
val maxes = booster_lags.select("max").as[String].collect()
val runSeq = (0 to params.length-1).map(ind => ParallelNotebookRunner("./time_align_parallel", 0, Map("device_name" -> params(ind), "max_interval" -> maxes(ind))))
val validation = parallelNotebooks(
  runSeq
)
Await.result(validation, 1440 minutes)
```
